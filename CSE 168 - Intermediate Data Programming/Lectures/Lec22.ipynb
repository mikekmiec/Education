# Image Demo
When we are working with images, it can be helpful to represent the image as a matrix.

```
import matplotlib.pyplot as plt
import numpy as np


import imageio
import matplotlib

def get_neighbors(i, j, img):
    vals = []
    if (i > 0):
        vals.append(img[i - 1, j])
    if (j > 0):
        vals.append(img[i][j - 1])
    if (i < len(img) - 1):
        vals.append(img[i + 1][j])
    if (j < len(img[0]) - 1):
        vals.append(img[i][j + 1])
    return [x for x in vals if x != 0]
```

```
img = imageio.imread('dog-corrupted.png')[:, :, 0]
plt.imshow(img, cmap=plt.cm.gray)
```

```
img
```

```
cuter = np.zeros((img.shape[0], img.shape[1]))
cuter
```

```
def smooth(img):
  cuter = np.zeros((img.shape[0], img.shape[1]))
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      if img[i, j] == 0:
        neighbors = get_neighbors(i, j, img)
        if len(neighbors) == 0:
          avg = 0
        else:
          avg = sum(neighbors) / len(neighbors)
        cuter[i][j] = avg
      else:
        cuter[i, j] = img[i, j]
      
  return cuter
```

We smooth the image twice to eliminate as many of the dots as possible.

```
cuter = smooth(img)
cutest = smooth(cuter)
plt.imshow(cutest, cmap=plt.cm.gray)
```
