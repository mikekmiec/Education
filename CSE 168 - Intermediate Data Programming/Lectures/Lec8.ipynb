```python
import pandas as pd
data = pd.read_csv('pokemon.csv')
data
```

# Seaborn Part 1
First we need to import seaborn and call the `set` function so it properly sets up all of the necessary state.

```python
import seaborn as sns
sns.set()
```

We first wanted to make a scatter plot of Attack vs Defense.

```python
sns.relplot(x="Attack", y="Defense", data=data)
```

We saw from the documentation that there are other parameters we can specify like the size of the dots based on another column.

```python
sns.relplot(x="Attack", y="Defense", size="Stage", data=data)
```

We then looked at another plot type that involves plotting a line for linear regression on the scatter plot. It shows uncertainty in the linear model with the band of blue showing potential lines that also fit this data with high probability. 

```python
sns.regplot(x='Attack', y='Defense', data=data)
```

# Seaborn Part 2
We wanted to then make a bar plot that counted up the number of each type of pokemon. We tried the default bar plot at first

```python
sns.catplot(x='Type 1', kind='count', data=data)
```

This plot has one big problem and one minor one. 
1. You can't read the x-axis which violates the major rule of making your visualization readable.
2. It's not necessarily a bad idea to encode the bars with different colors to emphasize they are different types (in fact, it sometimes can help by encoding information twice), but in this case, it doesn't necessarily make anything clearer.

We first fix the color issue

```python
sns.catplot(x='Type 1', kind='count', color='b', data=data)
```

Then we need to fix the axis issue. I'm sure there is a way to do this with just `seaborn`, but it seems pretty difficult to modfiy the chart through `seaborn`'s interface. This is generally where you bring `matplotlib` into the equation to help you  modify the current plot. 

On your homework, you should not need to make any of these adjustments with `matplotlib`, but you may need it to make your graphs for the project more legible. 

```python
import matplotlib.pyplot as plt
```

```python
sns.catplot(x='Type 1', kind='count', color='b', data=data)
plt.xticks(rotation=-45)
plt.savefig('bar_chart.png')
```

`matplotlib` lets us modify the last plot created in the way we want so we can make the graph more readable. 

Notice we also use `matplotlib` to save the plot to a file which is something you will need to do on the assignment.
